# Training and forecast example of an automated ARIMA model (TBATS).

```python
import numpy as np
import pandas as pd
from datetime import datetime
from sklearn.metrics import mean_squared_error as mse 
from sklearn.metrics import mean_absolute_percentage_error as mape
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels
import pickle
from datetime import timedelta
from tbats import TBATS
from sklearn.preprocessing import StandardScaler
import pickle
import math
from evaluate_forecasts import simple_n_day_ahead_forecast
```

## Load dataset

```python
ts60 = pd.read_csv('../../RDN/Load Data (2018-2019)/artifacts/load_60min.csv', index_col=0, parse_dates=True)
load60 = ts60['Load'].dropna()
load60.head()
```

## Seasonality detection
The time series is highly seasonal so this requires some further inspection.

```python
fig, ax = plt.subplots(2,1, figsize=(15, 8))
fig = sm.graphics.tsa.plot_acf(load60, lags=50, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(load60, lags=50, ax=ax[1])
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(load60,  model='additive', period=24)
f = result.plot()
f.set_figheight(15)
f.set_figwidth(25)
plt.show()
```

Daily seasonality is obvious from all above plots. e.g. ACF is repeated every 24 timesteps. Therefore seasonal models are expected to produce better results.

## Configuration

Steps for day ahead forecasts are preset to 24 (1 day). days_ahead are the number of days to reproduce forecast without retraining.

```python
steps = 24
days_ahead = 7
last_train_day = datetime(2019, 11, 30) # a week finishes there
model_id_date = '1_12'
```

## Train / test split

```python
from rdn_train_test_split import rdn_train_test_split
train, test, train_std, test_std, scalert = rdn_train_test_split(load60,
                                                                 last_train_day, 
                                                                 days_ahead,
                                                                 steps,
                                                                 freq='H', 
                                                                 scaler=StandardScaler())
```

## Training TBATS
In this section the TBATS model is fitted.

### Fit TBATS

```python
%%time
# Fit the model
estimator = TBATS(seasonal_periods=(24, 168, 365), n_jobs=8, use_box_cox=True, use_arma_errors=True, use_damped_trend=True, use_trend=True)
tbats = estimator.fit(train_std)
tbats.summary()
```

### Save model locally

```python
%%time
fname = f"../../RDN/Load Data (2018-2019)/models/rdn_tbats_{model_id_date}.pkl"  
with open(fname, 'wb') as file:  
    pickle.dump(tbats, file)
```

### Forecast

```python
simple_n_day_ahead_forecast(tbats, days_ahead, steps, train, test, StandardScaler())
```

## SARIMA

### Full SARIMA model from r gridsearch
A sarima is fitted as selected in the gridsearch process of R_tal_time_series_analysis.ipynb (further analysis can be sought there). However we have limited the seasonal autoregressive terms to 1 for speed and simplicity as longs as the model could not be saved otherwise. AIC, residuals, ACF, PACFs are presented again for completeness.

```python
%%time
sarima = sm.tsa.statespace.SARIMAX(endog=train_std, order=(2, 1, 1),
                                   seasonal_order=(4, 1, 2, 24)).fit(max_iter=200, method='powell')
print(sarima.summary())
```

```python
res = sarima.resid
fig, ax = plt.subplots(2,1, figsize=(20, 20))
fig = sm.graphics.tsa.plot_acf(res, lags=100, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(res, lags=100, ax=ax[1])
plt.show()
```

#### Simple n-day ahead forecast

```python
simple_n_day_ahead_forecast(sarima, days_ahead, steps, train, test, StandardScaler())
```

Ljung-Box seems pretty fine so we do not reject the null hypothesis of non-correlated residuals. However, the Jarque-Bera test results to rejection of the null hypothesis that the residual distribution has no skewness and kurtosis. Normality condition is not met at all...

#### Save model locally

```python
# workaround to make pickle work for SARIMAX
%%time
def __getnewargs__(self):
    return (tuple(i for i in self.params_complete))

statsmodels.api.tsa.statespace.SARIMAX.__getnewargs__ = __getnewargs__

# storing
fname = f"../../RDN/Load Data (2018-2019)/models/rdn_sarima_{model_id_date}.pkl"  
file = open(fname, 'wb')
pickle.dump(sarima, file)
file.close()
```

### Parsimonious SARIMA

```python
%%time
sarima_pars = sm.tsa.statespace.SARIMAX(endog=train_std, order=(2, 1, 1),
                                        seasonal_order=(2, 1, 2, 24)).fit(max_iter=200, method='powell')
print(sarima_pars.summary())
```

#### Simple n-day ahead forecast

```python
%%time
simple_n_day_ahead_forecast(sarima_pars, days_ahead, steps, train, test, StandardScaler())
```

#### Save model locally

```python
%%time
# workaround to make pickle work for SARIMAX
def __getnewargs__(self):
    return (tuple(i for i in self.params_complete))

statsmodels.api.tsa.statespace.SARIMAX.__getnewargs__ = __getnewargs__

# storing
fname = f"../../RDN/Load Data (2018-2019)/models/rdn_sarima_pars_{model_id_date}.pkl"  
file = open(fname, 'wb')
pickle.dump(sarima_pars, file)
file.close()
```

It is obvious that the model fails to capture the fluctuations of the inter-day variance of the load as all spikes (especially downward ones) cannot be forecasted effectively. This is due to the condition of constant variance in ARIMA models.

## SARIMAX
Gridsearch has run in R code. External variables are added to better capture complex seasonal relationships. Time variables are expected to help capture here more complex seasonalities and time relations. For that reason the parsimonious sarima will be used. Model is stored to disk. Weather is not yet available for RDN timeseries.

```python
# weather = pd.read_csv('../../RDN/Load Data (2018-2019)/artifacts/weather_curated_60min.csv',
#                       index_col=0, parse_dates=True)[datetime(2020, 11, 6):]
# weather.head()
```

```python
time = pd.read_csv('../../RDN/Load Data (2018-2019)/artifacts/time_60min.csv', index_col=1, parse_dates=True)
time_useful = time[['month_sin', 'month_cos', 'weekday_sin', 'weekday_cos', 
                    'hour_sin', 'hour_cos', 'holiday']].astype(int)
time_useful.head()
```

```python
# exog_training = pd.concat((weather.loc[train.index], time_useful.loc[train.index]), axis=1)
# exog_testing = pd.concat((weather.loc[test.index], time_useful.loc[test.index]), axis=1)
# exog_training.head()
exog_training = time_useful.loc[train.index]
exog_testing = time_useful.loc[test.index]
exog_training.head()
```

### Parsimonious model

```python
%%time
sarimax_pars = sm.tsa.statespace.SARIMAX(endog=train_std, 
                                         order=(2, 1, 1), 
                                         exog=exog_training.values,
                                         seasonal_order=(2, 1, 2, 24)) \
                                .fit(max_iter=200, method='powell')
print(sarimax_pars.summary())
```

#### Simple n-day ahead forecast
No need for scaler in exogenous variables. There are already scaled from preprocessing stage. 

```python
%%time
simple_n_day_ahead_forecast(sarimax_pars, 
                            days_ahead, 
                            steps, 
                            train, 
                            test, 
                            StandardScaler(), 
                            exog_train=exog_training, 
                            exog_test=exog_testing,
                            None)
```

#### Save model to disk

```python
%%time
def __getnewargs__(self):
    return (tuple(i for i in self.params_complete))

statsmodels.api.tsa.statespace.SARIMAX.__getnewargs__ = __getnewargs__

# storing
fname = f"../../RDN/Load Data (2018-2019)/models/rdn_sarimax_time_pars_{model_id_date}.pkl"
f = open(fname, 'wb')
pickle.dump(sarimax_pars, f)
f.close()
```
